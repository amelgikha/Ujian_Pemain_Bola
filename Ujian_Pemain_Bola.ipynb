```python
import numpy as np
import pandas as pd

df = pd.read_csv('data.csv')
df.head()
```

# Pemain yang Anda targetkan untuk direkrut memiliki kriteria sebagai berikut:

 - Usia (Age) <= 25 tahun
 - Skill umum (Overall) >= 80 point
 - Potensi (Potential) >= 80 point

```python
df1 = df[(df['Age']<=25)&(df['Overall']>=80)&(df['Potential']>=80)]
df1
```

# Gunakan minimal 3 buah algoritma model machine learning (pilihan model bebas), lalu bandingkan & tentukan model mana yang memiliki akurasi terbaik

```python
df['target'] = df.apply(
    lambda x : 1 if(x['Age']>=25) and (x['Overall']>=80) and (x['Potential']>=80) else 0, axis=1
)
df.head()
```

# Gunakan minimal 3 buah algoritma model machine learning (pilihan model bebas)

### Logistic Regression

```python
from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts = train_test_split(
    df[['Age','Overall','Potential']],
    df['target'],
    test_size = .05
)

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs')
model.fit(xtr,ytr)
model.score(xts,yts)
```

### Decision Tree

```python
from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts = train_test_split(
    df[['Age','Overall','Potential']],
    df['target'],
    test_size = .05
)

from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(xtr,ytr)
model.score(xts,yts)
```

### Random Forest

```python
from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts = train_test_split(
    df[['Age','Overall','Potential']],
    df['target'],
    test_size = .05
)

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, max_depth=2)
model.fit(xtr,ytr)
model.score(xts,yts)
```

# Menentukan model menggunakan k-fold validation

```python
from sklearn.model_selection import cross_val_score

print('Logistic Regression :', cross_val_score(
    LogisticRegression(solver='lbfgs'), xtr, ytr, cv=4
))
print('Decision Tree :', cross_val_score(
    tree.DecisionTreeClassifier(), xtr, ytr, cv=4
))
print('Random Forest :', cross_val_score(
    RandomForestClassifier(n_estimators=100, max_depth=2), xtr, ytr, cv=4
))
```
